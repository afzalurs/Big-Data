```python
import geopy
import pandas as pd
from geopy.geocoders import Nominatim
import time
from pprint import pprint
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="example app")
```

```python
import io
df2 = pd.read_csv('/content/students.csv')
df2
```

```python
df2.columns
```

```python
#concatenate address columns into one that is appropriate for geocoding

df2['Address']=df2['City'] + ' ,' + df2['District']+ ' ,' + df2['State']
```

```python
from geopy.extra.rate_limiter import RateLimiter
# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
# 2- - create location column
df2['location'] = df2['City'].apply(geocode)
```

```python
df2
```

```python
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
df2['point'] = df2['location'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
# 4 - split point column into latitude, longitude and altitude columns
df2[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df2['point'].tolist(), index=df2.index)
```

```python
df2.head()
```

```python
df2.columns
```

```python
#drop the unwanted columns
df2 = df2.drop(['Name', 'City', 'District', 'State', 'Address', 'location', 'point'], axis=1)
```

```python
#check the droped columns
df2.head()
```

```python
df2=df2.dropna(subset=['latitude'])
df2=df2.dropna(subset=['longitude'])
```

```python
#Creating a map of given locations
import folium
map1 = folium.Map(
    location=[59.338315,18.089960],
    tiles='cartodbpositron',
    zoom_start=12,
)
df2.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]]).add_to(map1), axis=1)
map1
```

```python
# Creating the Dark Map with Fast Marker Cluster

from folium.plugins import FastMarkerCluster
folium_map = folium.Map(location=[59.338315,18.089960],
                        zoom_start=20,
                        tiles='CartoDB dark_matter')


FastMarkerCluster(data=list(zip(df2['latitude'].values, df2['longitude'].values))).add_to(folium_map)
folium.LayerControl().add_to(folium_map)
folium_map
```

```python
#saving the map as html
folium_map.save("map2.html")
```

```python
# Importing the geodesic module from the library
from geopy.distance import geodesic

# Loading the lat-long data for Ghudda & Amritsar
Ghudda = (30.1244756, 74.78929509999999)
Amritsar = (31.6339793, 74.8722642)

# Print the distance calculated in km
print(geodesic(Ghudda, Amritsar).km)
```

```python
from geopy.distance import great_circle
print(great_circle(Ghudda, Amritsar).km)
```

```python
# import gmplot package
import gmplot

latitude_list = [ 30.3358376, 30.337977, 30.3316419 ]
longitude_list = [ 77.8701919, 78.848457, 78.8413095 ]

gmap3 = gmplot.GoogleMapPlotter(30.3358376,
								77.8701919, 13)

# scatter method of map object
# scatter points on the google map
gmap3.scatter( latitude_list, longitude_list, '# FF0000',
							size = 40, marker = False )

# Plot method Draw a line in
# between given coordinates
gmap3.plot(latitude_list, longitude_list,
		'cornflowerblue', edge_width = 2.5)

gmap3.draw( "map12.html" )
```
